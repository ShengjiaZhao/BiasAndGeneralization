```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def plot_batch(images):
    plt.figure(figsize=(6, 6))
    for i in range(0, 16):
        plt.subplot(4, 4, i+1)
        plt.imshow(images[i])
        plt.gca().xaxis.set_visible(False)
        plt.gca().yaxis.set_visible(False)
    plt.tight_layout()
    plt.show()
```

```python
from dataset_dots import DotsDataset
data_path = '<dataset_folder>'
dataset = DotsDataset(db_path=[os.path.join(data_path, '3_dots'), os.path.join(data_path, '6_dots')])
images = dataset.next_batch(100)
plot_batch(images)
```

```python
from dataset_pie import PieDataset
dataset = PieDataset(params=['40000','45555'])
images = dataset.next_batch(100)
plot_batch(images)
```
